```python
def age_convert(ages):
    age_categories = []
    for age in ages:
        if age < 20:
            age_categories.append("< 20_years")
        elif age < 26:
            age_categories.append("21-25_years")
        elif age < 31:
            age_categories.append("26-30_years")
        elif age < 36:
            age_categories.append("31-35_years")
        elif age < 41:
            age_categories.append("36-40_years")
        elif age < 46:
            age_categories.append("41-45_years")
        elif age < 51:
            age_categories.append("46-50_years")
        elif age < 56:
            age_categories.append("51-55_years")
        elif age < 61:
            age_categories.append("56-60_years")
        elif age < 66:
            age_categories.append("61-65_years")
        else:
            age_categories.append("> 65_years")
    return age_categories
```

```python
loan_default["age_groups"] = age_convert(loan_default["Age"])
```

```python
def Credit_score_convert(credit_score):
    credit_score_categories = []
    for i in credit_score:
        if i <= 350:
            credit_score_categories.append("< 350_score")
        elif i < 401:
            credit_score_categories.append("350-400_score")
        elif i < 451:
            credit_score_categories.append("401-450_score")
        elif i < 501:
            credit_score_categories.append("451-500_score")
        elif i < 551:
            credit_score_categories.append("501-550_score")
        elif i < 601:
            credit_score_categories.append("551-600_score")
        elif i < 651:
            credit_score_categories.append("601-650_score")
        elif i < 701:
            credit_score_categories.append("651-700_score")
        elif i < 751:
            credit_score_categories.append("701-750_score")
        elif i < 801:
            credit_score_categories.append("751-800_score")
        elif i < 851:
            credit_score_categories.append("801-850_score")
        else:
            credit_score_categories.append("> 851_score")
    return credit_score_categories
```

```python
loan_default["credit_Score_group"] = Credit_score_convert(loan_default["CreditScore"])
```

```python
loan_default.head(10)
```

```python
def loanterm_convert(LoanTerm):
    LoanTerm_categories = []
    for i in LoanTerm:
        if i < 12:
            LoanTerm_categories.append("< 12_months")
        elif i < 19:
            LoanTerm_categories.append("12-18_months")
        elif i < 26:
            LoanTerm_categories.append("19-25_months")
        elif i < 33:
            LoanTerm_categories.append("26-32_months")
        elif i < 40:
            LoanTerm_categories.append("33-39_months")
        elif i < 47:
            LoanTerm_categories.append("40-46_months")
        elif i < 54:
            LoanTerm_categories.append("47-53_months")
        elif i > 53:
            LoanTerm_categories.append(">53_months")
    return LoanTerm_categories
```

```python
loan_default["Loan_term_Group"]=loanterm_convert(loan_default["LoanTerm"])
```
